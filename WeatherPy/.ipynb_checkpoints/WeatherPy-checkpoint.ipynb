```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "C:/Users/prakb/OneDrive/Desktop/python-api-challenge/output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
url = "https://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
Cities = []
Clouds = []
Dates = []
Humidities = []
Latitudes = []
Longitudes = []
Max_Temperature = []
Wind_Speeds = []
Countries = []
```

```python
city_number = 0
for city in cities:
    try:
      
        query_url = f"{url}appid={weather_api_key}&q={city}&units={units}"
        WeatherPy_json = response = requests.get(query_url).json()
        city_counter = city_counter + 1
        print(city, city_counter)
    
        var1 = (WeatherPy_json["name"])
        var2 = (WeatherPy_json["clouds"]["all"])
        var3 = (WeatherPy_json["dt"])
        var4 = (WeatherPy_json["main"]["humidity"])
        var5 = (WeatherPy_json["coord"]["lat"])
        var6 = (WeatherPy_json["coord"]["lon"])
        var7 = (WeatherPy_json["main"]["temp_max"])
        var8 = (WeatherPy_json["wind"]["speed"])
        var9 = (WeatherPy_json["sys"]["country"])
        
        Cities.append(var1)
        Clouds.append(var2)
        Dates.append(var3)
        Humidities.append(var4)
        Latitudes.append(var5)
        Longitudes.append(var6)
        Max_Temperature.append(var7)
        Wind_Speeds.append(var8)
        Countries.append(var9)
    except:
        print("skipped")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
WeatherPy_df = pd.DataFrame({
    "City": Cities,
    "Clouds": Clouds,
    "Dates": Dates,
    "Humidity": Humidities,
    "Latitude": Latitudes,
    "Longitude": Longitudes,
    "Temperature":Max_Temperature,
    "Wind Speed": Wind_Speeds,
    "Country":Countries,
})
WeatherPy_df.head()
```

```python
WeatherPy_df.to_csv(output_data_file)
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
LatTemp2 = plt.scatter(Northern["Latitude"], Northern["Temperature"], marker = "o", facecolors = "blue", edgecolors = "Black")
plt.xlabel("Latitude")
plt.ylabel("Temperature (F)")
plt.grid()
plt.title("Latitude vs Max Temperature (F) (Northern Hemisphere)")
(slope, intercept, rvalue, pvalue, stderr) = linregress(Northern["Latitude"], Northern["Temperature"])
regression1 = Northern["Latitude"] * slope + intercept
slope1 = 'y = ' + str(round(slope,2)) + 'x +' + str(round(intercept,2))
plt.plot(Northern["Latitude"], regression1, 'Red')
plt.annotate(slope1,(0,45), fontsize=14, color='black')
plt.show()
```
